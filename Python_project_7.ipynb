```python
import pandas as pd

df = pd.read_csv('covid.csv')

df.head()
```

```python

print(df.columns)
df.info()
df.describe()
```

```python

top10 = df.sort_values(by='Confirmed', ascending=False).head(10)

print(top10[['Country/Region', 'Confirmed']])
```

```python
import matplotlib.pyplot as plt

top10.plot(x='Country/Region', y='Confirmed', kind='bar')

plt.title('Top 10 Countries - COVID Confirmed Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')

plt.show()
```
